```python
#Question 1

# Step 1: Import the necessary libraries
import pandas as pd

# Step 2: Import the dataset from "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')

# Step 3: Assign it to a variable called users
users = pd.read_csv(url, sep='|')

# Step 4: Discover what is the mean age per occupation
mean_age = users.groupby('occupation')['age'].mean()
print("\n===(Mean Age per Occupation)===\n")
print(mean_age)

# Step 5: Discover the Male ratio per occupation and sort it from the most to the least
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0

users['gender_n'] = users['gender'].apply(gender_to_numeric)
male_ratio = users.groupby('occupation')['gender_n'].sum() / users['occupation'].value_counts() * 100
male_ratio.sort_values(ascending=False, inplace=True)
print("\n===(Male Ratio per Occupation)===\n")
print(male_ratio)

# Step 6: For each occupation, calculate the minimum and maximum ages
min_age = users.groupby('occupation')['age'].min()
max_age = users.groupby('occupation')['age'].max()
print("\n===(Minimum Age per Occupation)===\n")
print(min_age)
print("\n===(Maximum Age per Occupation)===\n")
print(max_age)

# Step 7: For each combination of occupation and sex, calculate the mean age
mean_age_occ_sex = users.groupby(['occupation', 'gender'])['age'].mean()
print("\n===(Mean Age per Occupation and sex)===\n")
print(mean_age_occ_sex)

# Step 8: For each occupation present the percentage of women and men
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occup_count = users.groupby(['occupation']).agg('count')
occup_gender = gender_ocup.div(occup_count, level = "occupation") * 100
print("\n===(Percentage of women and men per Occupation)===\n")
print(occup_gender.loc[:, 'gender'])
```

```python
# Step #Question 2: Import the necessary libraries
import pandas as pd

# Step 2: Import the dataset from "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(url)

# Step 3: Assign it to a variable called euro12
euro12 = pd.read_csv(url)

# Step 4: Select only the Goal column
goals = euro12['Goals']
print("\n===(Goals only)===\n")
print(goals)

# Step 5: How many team participated in the Euro2012?
num_teams = len(euro12['Team'].unique())
print("\n===(Number of teams: )===\n")
print(num_teams)

# Step 6: What is the number of columns in the dataset?
num_cols = len(euro12.columns)
print("\n===(Number of Columns in the dataset: )===\n")
print(num_cols)

# Step 7: View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("\n===( Discipline )===\n")
print(discipline)

# Step 8: Sort the teams by Red Cards, then to Yellow Cards
discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending=[False, False], inplace=True)
print("\n===(Teams by Red and Yellow cards)===\n")
print(discipline)

# Step 9: Calculate the mean Yellow Cards given per Team
mean_ycards = euro12['Yellow Cards'].mean()
print("\n===(Mean Yellow cards)===\n")
print(mean_ycards)

# Step 10: Filter teams that scored more than 6 goals
high_scoring_teams = euro12[euro12['Goals'] > 6]
print("\n===(Teams that scored more than 6 goals )===\n")
print(high_scoring_teams)

# Step 11: Select the teams that start with G
G_teams = euro12[euro12['Team'].str.startswith('G')]
print("\n===(Teams that start with G )===\n")
print(G_teams)

# Step 12: Select the first 7 columns
first_7_cols = euro12.iloc[:, :7]
print("\n===( First 7 columns )===\n")
print(first_7_cols)

# Step 13: Select all columns except the last 3
all_except_last_3_cols = euro12.iloc[:, :-3]
print("\n===(All columns except the last 3 )===\n")
print(all_except_last_3_cols)

# Step 14: Present only the Shooting Accuracy from England, Italy and Russia
SAs = euro12.loc[euro12['Team'].isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']]
print("\n===( Shooting accuracy of England, Italy and Russia )===\n")
print(SAs)
```

```python
#Question 4

# Step 1: Import necessary libraries
import numpy as np
import pandas as pd

# Step 2: Import dataset
data = pd.read_csv('wind.txt', sep='\s+', parse_dates=[[0,1,2]])

# Step 3: Set datetime index
data = data.set_index('Yr_Mo_Dy')

# Step 4: Function to fix year 2061
def fix_year(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return pd.to_datetime(year*10000 + x.month*100 + x.day, format='%Y%m%d')

# Step 5: Apply function to fix year 2061
data.index = data.index.map(fix_year)

# Step 6: Compute number of missing values for each location
print("\n===( Number of Missing values for each Location )===\n")
print("Missing values per location:\n", data.isna().sum())

# Step 7: Compute number of non-missing values
print("\n===( Number of Non-Missing values  )===\n")
print("Total non-missing values:", data.notna().sum().sum())

# Step 8: Compute mean windspeed over all locations and times
print("\n===( Mean Windspeed over all Location and Time )===\n")
print("Mean windspeed:", data.mean().mean())

# Step 9: Compute min, max, mean, and standard deviation of windspeed for each location
loc_stats = data.agg(['min', 'max', 'mean', 'std'], axis=0).transpose()
print("\n===( Min, Max, Mean, and Standard Deviation of Windspeed for each Location )===\n")
print("Location statistics:\n", loc_stats)

# Step 10: Compute min, max, mean, and standard deviation of windspeed for each day
day_stats = data.agg(['min', 'max', 'mean', 'std'], axis=1)
print("\n===( Min, Max, Mean, and Standard Deviation of Windspeed for each Day )===\n")
print("Day statistics:\n", day_stats)

# Step 11: Compute mean windspeed in January for each location
jan_data = data[data.index.month == 1]
jan_stats = jan_data.mean()
print("\n===( Mean Windspeed in January for each Location )===\n")
print("January windspeed by location:\n", jan_stats)

# Step 12: Downsample to yearly frequency for each location
yearly_data = data.resample('Y').mean()
print("\n===( Downsampled yearly Frequency for each Location )===\n")
print("Yearly data:\n", yearly_data)

# Step 13: Downsample to monthly frequency for each location
monthly_data = data.resample('M').mean()
print("\n===( Downsampled Monthly Frequency for each Location )===\n")
print("Monthly data:\n", monthly_data)

# Step 14: Downsample to weekly frequency for each location
weekly_data = data.resample('W').mean()

print("\n===( Downsampled Weekly Frequency for each Location )===\n")
print("Weekly data:\n", weekly_data)

# Step 15: Compute min, max, mean, and standard deviation of windspeed for each week of the first 52 weeks
weekly_stats = weekly_data.iloc[:52].agg(['min', 'max', 'mean', 'std'], axis=1)
print("\n===( Min, Max, Mean, and Standard Deviation of Windspeed for each Week of first 52 weeks )===\n")
print("Weekly statistics for first 52 weeks:\n", weekly_stats)

```

```python
#Question 5

# Step 1. Import the necessary libraries
import pandas as pd

# Step 2. Import the dataset
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, delimiter = '\t')

# Step 3. Assign it to a variable called chipo

# Step 4. See the first 10 entries
print("\n===( First 10 entries )===\n")
print(chipo.head(10))

# Step 5. What is the number of observations in the dataset?
print("\n===( Number of observations in the dataset )===\n")
print('Number of observations:', len(chipo))

# Step 6. What is the number of columns in the dataset?
print("\n===( Number of Columns in the dataset )===\n")
print('Number of columns:', len(chipo.columns))

# Step 7. Print the name of all the columns.
print("\n===( Name of all the columns )===\n")
print('Columns:', chipo.columns)

# Step 8. How is the dataset indexed?
print("\n===( The dataset is indexed as: )===\n")
print('Index:', chipo.index)

# Step 9. Which was the most-ordered item?
item_count = chipo.groupby('item_name')['quantity'].sum()
most_ordered_item = item_count.idxmax()
print("\n===( The most ordered item: )===\n")
print('Most ordered item:', most_ordered_item)

# Step 10. For the most-ordered item, how many items were ordered?
print("\n===( Number of ordered items for the most ordered item: )===\n")
print('Quantity ordered:', item_count[most_ordered_item])

# Step 11. What was the most ordered item in the choice_description column?
choice_count = chipo.groupby('choice_description')['quantity'].sum()
most_ordered_choice = choice_count.idxmax()
print("\n===( The most ordered item in the choice_description column: )===\n")
print('Most ordered choice:', most_ordered_choice)

# Step 12. How many items were ordered in total?
total_items_ordered = chipo['quantity'].sum()
print("\n===( Total ordered item: )===\n")
print('Total items ordered:', total_items_ordered)

# Step 13.
print("\n===( Step 13 )===\n")
# Turn the item price into a float
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)

# Check the item price type
print('Type of item price:', chipo['item_price'].dtype)

# Create a lambda function and change the type of item price
chipo['item_price'] = chipo['item_price'].apply(lambda x: '{:.2f}'.format(x)).astype(float)

# Check the item price type
print('Type of item price after lambda function:', chipo['item_price'].dtype)

# Step 14. How much was the revenue for the period in the dataset?
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("\n===( Revenue for the period in the dataset: )===\n")
print('Revenue:', revenue)

# Step 15. How many orders were made in the period?
orders = chipo['order_id'].nunique()
print("\n===( Total orders for the period in the dataset: )===\n")
print('Number of orders:', orders)

# Step 16. What is the average revenue amount per order?
average_revenue_per_order = revenue / orders
print("\n===( Average Revenue per Order: )===\n")
print('Average revenue per order:', average_revenue_per_order)

# Step 17. How many different items are sold?
unique_items_sold = chipo['item_name'].nunique()
print("\n===( Different Items sold )===\n")
print('Number of different items sold:', unique_items_sold)
```

```python
#Question 6

import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv("us-marriages-divorces-1867-2014.csv", index_col=0)

# Create a line plot of marriages and divorces per capita
plt.plot(data.index, data["Marriages_per_1000"], label="Marriages")
plt.plot(data.index, data["Divorces_per_1000"], label="Divorces")

# Add axis labels and a title
plt.xlabel("Year")
plt.ylabel("Marriages and Divorces per 1000 people")
plt.title("Marriages and Divorces per Capita in the US (1867-2014)")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
#Question 7

import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv("us-marriages-divorces-1867-2014.csv")

# Filter the data to include only the years we're interested in
years = [1900, 1950, 2000]
filtered_data = data[data['Year'].isin(years)]

# Create a bar chart
plt.bar(filtered_data['Year'] - 5, filtered_data['Marriages_per_1000'], width=5, label='Marriages')
plt.bar(filtered_data['Year'], filtered_data['Divorces_per_1000'], width=5, label='Divorces')

# Add labels to the chart
plt.xlabel('Year')
plt.ylabel('Marriages and Divorces per 1000')
plt.title('Marriages and Divorces per Capita in the U.S. (1900, 1950, and 2000)')
plt.xticks(years)
plt.legend()

# Display the chart
plt.show()

```

```python
#Question 9

import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file into a DataFrame
data = pd.read_csv('roman-emperor-reigns.csv')

# Count the number of emperors who were assassinated
assassinated_count = len(data[data['Cause_of_Death'] == 'Assassinated'])

# Count the number of emperors who were not assassinated
not_assassinated_count = len(data) - assassinated_count

# Create a list of labels for the pie chart
labels = ['Assassinated', 'Not Assassinated']

# Create a list of values for the pie chart
values = [assassinated_count, not_assassinated_count]

# Create a pie chart with the specified labels, values, and formatting
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')

# Add a title to the pie chart
plt.title('Fraction of Roman Emperors who were Assassinated')

# Show the pie chart
plt.show()
```
